## Pearson's correlation

```python
corr_matrix = pd.DataFrame(df.corr()['Stability'].sort_values())

corr_matrix.drop('Stability', axis=0, inplace=True) # Drop row stability

corr_matrix
```

Some features (104) have `NaN` correlation with `Stability`. Why?

### `NaN` as correlation with `Stability`

```python
corr_matrix[corr_matrix['Stability'].isna()]
```

```python
nan_corr = pd.DataFrame()

nan_corr['mean'] = df[corr_matrix[corr_matrix['Stability'].isna()].index.values].describe().loc['mean']
nan_corr['std'] = df[corr_matrix[corr_matrix['Stability'].isna()].index.values].describe().loc['std']
nan_corr['min'] = df[corr_matrix[corr_matrix['Stability'].isna()].index.values].describe().loc['min']
nan_corr['max'] = df[corr_matrix[corr_matrix['Stability'].isna()].index.values].describe().loc['max']
nan_corr.sort_values('mean')
```

```python
nan_corr[nan_corr['min']==nan_corr['max']].sort_values('mean')
```

103 of these features have constant value everytime.

Details show below:

```python
nan_corr[nan_corr['min']==nan_corr['max']]['mean'].value_counts()
```

```python
i_c = df.columns[df.nunique() <= 1]
df[i_c].describe().T.sort_values('mean')
```

```python
nan_corr[nan_corr['min']!=nan_corr['max']]
```

Add above to rules..
